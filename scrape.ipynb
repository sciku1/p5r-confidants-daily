```python
import pandas as pd
from selenium import webdriver
 
 
driver = webdriver.Firefox()
driver.get("https://gamefaqs.gamespot.com/ps4/260936-persona-5-royal/faqs/78212/confidant-availability")

tables = driver.find_elements_by_css_selector("#faqwrap table")

dfs = []
for tbl in tables:
    table_html = tbl.get_attribute("outerHTML")
    parsed_table = pd.read_html(table_html)[0]
    dfs.append(parsed_table)


# easy.head()
driver.close()
```

```python
dataframe = pd.concat(dfs)
 
dataframe.reset_index(inplace=True)
dataframe.drop(0)

data = dict()
for index, row in dataframe.iterrows():
    
    if data.get(row['Date'], None) is None: 
        data[row['Date']] = list()
    for colname, colval in row.iteritems():
       if colname not in ["index", "Day", "Date"]:
           if colval == "O":
                data[row['Date']].append(colname)

filtered_data = {k:v for k,v in data.items() if v}
filtered_data
```

```python
from json import dumps
target_file = "./src/data/data.json"

with open(target_file, "w") as file:
    file.write(dumps(filtered_data))
```
